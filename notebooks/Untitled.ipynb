```python
import ROOT
import pandas as pd
import numpy as np
import os
import sys
import ipynbname
from pathlib import Path

project_root = str(ipynbname.path().parent.parent)
sys.path.append(project_root)

from processing import SiphraAcquisition
from analysis import fit_peak_expbg
```

```python
# Constants
BITS12 = 2**12
BITS9 = 2**9 # 512 typical number of bins used
# Energy emission peaks in MeV
K40_MEV = 1.460
TL208_MEV = 2.614
CS137_MEV = 0.661
colors = [ROOT.kRed, ROOT.kBlue, ROOT.kGreen, ROOT.kOrange, ROOT.kViolet, ROOT.kYellow, ROOT.kSpring, ROOT.kCyan,]
```

```python
channels = [f'Ch{n}' for n in range(2,17,2)]
files_lsts = {}
files_lsts[channels[0]] = sorted(Path(project_root+'/data/260311').glob('SUBTRACTED*Ch2*.csv'))
files_lsts[channels[0]].pop(3)
files_lsts[channels[0]].pop(3)
files_lsts[channels[0]].pop(3)
for ch in channels[1:]:
    files_lsts[ch] = sorted(Path(project_root+'/data/260311').glob(f'SUBTRACTED*{ch}*.csv'))

vbiases = [28.0, 28.5, 29.0]
# Datasets
datasets = {}
for ch, lst in files_lsts.items():
    datasets[ch] = []
    datasets[ch].append(SiphraAcquisition(lst[2]))
    datasets[ch].append(SiphraAcquisition(lst[0]))
    datasets[ch].append(SiphraAcquisition(lst[1]))
del(files_lsts)
```

```python
if ROOT.gROOT.FindObject('cv'):
    ROOT.gROOT.FindObject('cv').Close()

cv = ROOT.TCanvas('cv', 'cv', 2000, 1200)
cv.divide(4,2)
ROOT.gStyle.SetOptStat(11)
ROOT.gStyle.SetStatFontSize(0.03)
ROOT.gStyle.SetStatW(0.16)

hists = {}
legends = {}

for idx, (ch, dataset) in enumerate(datasets.items()):
    cv.cd(idx)
    hists[ch] = []
    lg = ROOT.TLegend(0.48, 0.61, 0.75, 0.83)
    for acq, vbias in zip(dataset, vbiases):
        hist = ROOT.TH1F(f"{ch} Vbias={vbias}", "", BITS12, 0, BITS12)
        hist.Fill(acq[ch])
        hist.Scale(1/acq.exposure)
        hists[ch].append(hist)
        hist.GetXaxis().SetTitle("ADC channel number")
        hist.GetYaxis().SetTitle(r"Counts")
        hist.SetLineColor(colors[idx])
        del(hist)
        

        
```
