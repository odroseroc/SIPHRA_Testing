```python
if ROOT.gROOT.FindObject("c1"):
    ROOT.gROOT.FindObject("c1").Close()

c1 = ROOT.TCanvas("c1", "c1", 800, 600)
ROOT.gStyle.SetOptStat(True)
ROOT.gStyle.SetStatFontSize(0.03)
ROOT.gStyle.SetStatW(0.15)

IamLegend = ROOT.TLegend(0.35, 0.67, 0.48, 0.81)

ch_spectra = []
frame_lst = dfs_SiPM14_BG
colors = [ROOT.kBlue, ROOT.kRed, ROOT.kGreen, ROOT.kOrange]
for df, ch, lg, t, c in zip(frame_lst, SIPHRA_Ch_list, legend, times_SiPM14_BG, colors):
    hist = ROOT.TH1F(ch, "Background spectra SiPM Ch. 1-4", N_BINS, 0, BITS_12)
    for dp in df[ch]:
        hist.Fill(dp)
    hist.Scale(1/t)
    hist.GetXaxis().SetTitle("ADC channel number")
    hist.GetYaxis().SetTitle(r"Count rate (s^{-1})")
    hist.SetLineColor(c-2)
    st=hist.FindObject("stats")
    # st.SetX1NDC(0.65)
    IamLegend.SetHeader('SIPHRA channel')
    IamLegend.AddEntry(hist, lg, 'l')
    ch_spectra.append(hist)

ch_spectra[0].Draw("hist")
for sp in ch_spectra[1:]:
    sp.Draw("hist,sames")
c1.SetLogy()
IamLegend.Draw()

c1.Draw()
ROOT.gPad.Modified()
ROOT.gPad.Update()
```

```python
def gauspeak(x, mean, sigma, norm):
    return norm*ROOT.TMath.Gaus(x, mean, sigma)
def bg_exp(x, p0, p1):
    return p0[0]*ROOT.Tmath.Exp(-x[0]/p1[0])
def totalfn(x, p1, mean, sigma, norm):
    return gb_exp(x, p1) + gauspeak(x, mean, sigma, norm)

formula = "[0]*TMath::Exp(-x/[1]) + [2]*TMath::Gaus(x, [3], [4])"

c2 = ROOT.TCanvas("c2", "c2", 800, 600)
# bg_exp = ROOT.TF1("ExpDecayBG", "[p0]*TMath::Exp(x*[p1]/[p2])", 0, BITS_12)
# gaussian1 = ROOT.TF1("K40peak", "TMath::Gaus(x, [mean], [sigma])", 800, 1100)
# gaussian2 = ROOT.TF1("Tl208peak", "TMath::Gaus(x, [mean], [sigma])", 1300, 1600)

bg = ROOT.TF1("ExpDecayBG", formula, 700, 1200, 5)
bg.SetParameters(180, 200, 4, 904, 4)

fitResult = ch_spectra[2].Fit(bg, "L S", "", 700, 1200)
ch_spectra[2].GetXaxis().SetRangeUser(0, BITS_12)
ch_spectra[2].Draw("SAME hist")
c2.SetLogy()
c2.Draw()
```

```python
import ROOT

# 1. Definimos la funcin TOTAL usando un string de C++
# [0] = p0 (Amplitud Fondo)
# [1] = p1 (Tau Fondo)
# [2] = Norm (Norma Pico)
# [3] = Mean (Media Pico)
# [4] = Sigma (Sigma Pico)
formula = "[0]*TMath::Exp(-x/[1]) + [2]*TMath::Gaus(x, [3], [4])"

# 2. Creamos el TF1 usando el string.
# El '5' al final indica que hay 5 parmetros entre corchetes []
func_total = ROOT.TF1("func_total", formula, 100, 400)

# 3. ESTO ES CRUCIAL: Dar valores iniciales cercanos a la realidad
# Si no pones valores iniciales, el ajuste diverge y a veces crashea
func_total.SetParameters(180, 200, 1000, 250, 50) 
# Ajusta estos valores: (Amp_BG, Tau_BG, Norm_Gaus, Mean_Gaus, Sigma_Gaus)

# Asumimos que ch_spectra[2] existe
c2 = ROOT.TCanvas("c2", "c2", 800, 600)

# Opciones de ajuste: "L" (Likelihood), "S" (Guardar resultado), "R" (Usar rango definido)
# Nota: Al pasar el objeto 'func_total', es ms seguro que pasar el nombre string
fitResult = ch_spectra[2].Fit(func_total, "L S R")

# Configuracin del dibujo
ch_spectra[2].GetXaxis().SetRangeUser(50, 450)
ch_spectra[2].Draw("SAME")

c2.Draw()
```
