```python
if ROOT.gROOT.FindObject("c1"):
    ROOT.gROOT.FindObject("c1").Close()

c1 = ROOT.TCanvas("c1", "c1", 800, 600)
ROOT.gStyle.SetOptStat(True)

IamLegend = ROOT.TLegend(0.35, 0.67, 0.45, 0.8)

ch_spectra = []
frame_lst = dfs_SiPM14_BG
colors = ["kBlue", "kRed", "kGreen", "kOrange"]
for df, ch, lg, t, c in zip(frame_lst, SIPHRA_Ch_list, legend, times_SiPM14_BG, colors):
    hist = ROOT.TH1F(ch, "Background spectra SiPM Ch. 1-4", N_BINS, 0, BITS_12)
    for dp in df[ch]:
        hist.Fill(dp)
    hist.Scale(1/t)
    hist.GetXaxis().SetTitle("ADC channel number")
    hist.GetYaxis().SetTitle(r"Count rate (s^{-1})")
    hist.SetLineColor(c)
    IamLegend.AddEntry(hist, lg)
    ch_spectra.append(hist)

ch_spectra[0].Draw("hist")
for sp in ch_spectra[1:]:
    sp.Draw("hist,sames")
c1.SetLogy()
IamLegend.Draw()

c1.Draw()
ROOT.gPad.Modified()
ROOT.gPad.Update()
```
